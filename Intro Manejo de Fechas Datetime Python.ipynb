##SEGUNDO EJEMPLO CON VALORES DE TIPO HORA

```python
dh=pd.read_csv('/Users/martinlopezscala/Downloads/Sales_April_2019.csv')
```

```python
dh.info()
```

```python
dh.head()
```

```python
#dh[dh['Order Date'].isnull()]
dh=dh[dh['Order Date'].notnull()]
```

```python
#Ver Unknown string format: Order Date present at position 0
#dh[dh['Order Date']=='Order Date']
dh=dh[dh['Order Date']!='Order Date']
```

```python
#dh['Order #Date']=dh['Order Date'].apply(lambda x: datetime.fromisoformat(str(x))) #Invalid isoformat string: '04/19/19 0
#dh['Order Date']= dh['Order Date'].apply(lambda x: datetime.fromisoformat(x)) #Invalid isoformat string: '04/19/19 08:46'
##dh['Order Date']= dh['Order Date'].apply(lambda x: pd.to_datetime(x)) #Unknown string format: Order Date present at position
dh['Order Date']= dh['Order Date'].apply(lambda x: pd.to_datetime(x))
```

```python
dh.head()
```

```python
#Order Date        18289 non-null  datetime64[ns]
dh.info()
```

```python
#Extraer el año
dh['YEAR']=dh['Order Date'].apply(lambda x: x.year)
dh['MONTH']=dh['Order Date'].apply(lambda x: x.month)
dh['DAY']=dh['Order Date'].apply(lambda x: x.day)
dh['HOUR']=dh['Order Date'].apply(lambda x: x.hour)
dh['MINUTE']=dh['Order Date'].apply(lambda x: x.minute)
```

```python
dh.head()
```

```python
#Fechas personalizadas para comparar
d=date(2019,4,19)
t=time(12,30)
fecha_contraste=datetime.combine(d,t)
fecha_contraste
```

```python
#Validar la hora
dh['VALIDAR HORA']=dh['Order Date'].apply(lambda x: x>fecha_contraste)
```
